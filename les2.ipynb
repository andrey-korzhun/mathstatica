## Урок 2. Дискретные случайные величины. Закон распределения вероятностей. Биномиальный закон распределения. Распределение Пуассона

```python
def factorial(n):
    if n < 2:
        return 1
    else:
        return n * factorial(n-1)

    
def e(n):
    e = 0
    for x in range(n):
        e += 1 / factorial(x)
    return e


def combinations(n, k):
    return int(factorial(n) / (factorial(k) * factorial(n - k)))


def bernoulli(n, k, p):
    return combinations(n,k) * (p ** k) * ((1 - p) ** (n - k))


def poisson(n, m, p):
    das_lambda = n * p
    return ((das_lambda ** m) / factorial(m)) * (e(8) ** - das_lambda)
```

### 1. Вероятность того, что стрелок попадет в мишень, выстрелив один раз, равна 0.8. Стрелок выстрелил 100 раз. Найдите вероятность того, что стрелок попадет в цель ровно 85 раз.

```python
round(bernoulli(100, 85, 0.8), 8)
```

### 4. а) В первом ящике находится 10 мячей, из которых 7 - белые. Во втором ящике - 11 мячей, из которых 9 белых. Из каждого ящика вытаскивают случайным образом по два мяча. Какова вероятность того, что все мячи белые? 

```python
ans = round(7/10 * 6/9 * 9/11 * 8/10, 8)
ans
```

#### Первый ящик

```python
C_7_2 = combinations(7, 2)
C_10_2 = combinations(10 , 2)
C_7_2, C_10_2
```

#### Второй ящик

```python
C_9_2 = combinations(9, 2)
C_11_2 = combinations(11, 2)
C_9_2, C_11_2
```

#### Итого

```python
ans_4a = round(C_7_2 / C_10_2 * C_9_2 / C_11_2, 8)
ans_4a
```

```python
ans == ans_4a
```

### 4. б) В первом ящике находится 10 мячей, из которых 7 - белые. Во втором ящике - 11 мячей, из которых 9 белых. Из каждого ящика вытаскивают случайным образом по два мяча. Какова вероятность того, что ровно два мяча белые?

```python
ans = round(7/10 * 6/9 * 2/11 * 1/10 +
            (7/10 * 3/9 + 3/10 * 7/9) *
            (9/11 * 2/10 + 2/11 * 9/10) +
            9/11 * 8/10 * 3/10 * 2/9,
            8)
ans
```

#### 2 белых из первого и 2 не белых из второго

```python
C_2_2 = combinations(2, 2)
C_7_2, C_10_2, C_2_2, C_11_2
```

#### 1 белый и 1 не белый из каждого

```python
C_7_1 = combinations(7, 1)
C_3_1 = combinations(3, 1)
C_9_1 = combinations(9, 1)
C_2_1 = combinations(2, 1)
C_11_2 = combinations(11, 2)
C_7_1, C_3_1, C_10_2, C_9_1, C_2_1, C_11_2
```

#### 2 не белых из первого и 2 белых из второго

```python
C_3_2 = combinations(3, 2)
C_3_2, C_10_2, C_9_2, C_11_2
```

#### Итого

```python
ans_4b = round(
    C_7_2 / C_10_2 * C_2_2 / C_11_2 +
    C_7_1 * C_3_1 / C_10_2 * C_9_1 * C_2_1 / C_11_2 +
    C_3_2 / C_10_2 * C_9_2 / C_11_2,
    8)
ans_4b
```

```python
ans == ans_4b
```
