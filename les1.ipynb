## Урок 1. Случайные события. Условная вероятность. Формула Байеса. Независимые испытания

```python
def factorial(n):
    if n < 2:
        return 1
    else:
        return n * factorial(n-1)
    

def combinations(n, k):
    return int(factorial(n) / (factorial(k) * factorial(n - k))) 
```

### 1. а) Из колоды в 52 карты извлекаются случайным образом 4 карты. Найти вероятность того, что все карты – крести.

```python
ans_1a = round(13/52 * 12/51 * 11/50 * 10/49, 8)
ans_1a
```

#### Проверка

```python
C_13_4 = combinations(13, 4)
C_13_4
```

```python
C_52_4 = combinations(52, 4)
C_52_4
```

```python
ans = round(C_13_4  / C_52_4, 8)
ans
```

```python
ans_1a == ans
```

### 1. б) Из колоды в 52 карты извлекаются случайным образом 4 карты. Найти вероятность, что среди 4-х карт окажется хотя бы один туз.

#### Рассмотрим подмножество без тузов

```python
C_48_4 = combinations(48, 4)
C_48_4
```

```python
C_4_ace = C_52_4 - C_48_4
C_4_ace
```

```python
ans_1b = round(C_4_ace / C_52_4, 8)
ans_1b
```

#### Проверка

```python
C_4_1 = combinations(4, 1)
C_4_1
```

```python
C_48_3 = combinations(48, 3)
C_48_3
```

```python
C_4_2 = combinations(4, 2)
C_4_2
```

```python
C_48_2 = combinations(48, 2)
C_48_2
```

```python
C_4_3 = combinations(4, 3)
C_4_3
```

```python
C_48_1 = combinations(48, 1)
C_48_1
```

```python
C_4_4 = combinations(4, 4)
C_4_4
```

```python
C_48_0 = combinations(48, 0)
C_48_0
```

```python
ans = round(
    (C_4_1*C_48_3 + C_4_2*C_48_2 + C_4_3*C_48_1 + C_4_4*C_48_0) / C_52_4, 8)
ans
```

```python
ans_1b == ans
```

### 2. На входной двери подъезда установлен кодовый замок, содержащий десять кнопок с цифрами от 0 до 9. Код содержит три цифры, которые нужно нажать одновременно. Какова вероятность того, что человек, не знающий код, откроет дверь с первой попытки?

```python
ans_2 = round(3/10 * 2/9 * 1/8, 8)
ans_2
```

#### Проверка

```python
ans_2 == ans
```
