## Урок 8. Дисперсионный анализ. Логистическая регрессия

```python
import numpy as np
from scipy import stats
```

### Провести дисперсионный анализ для определения того, есть ли различия среднего роста среди взрослых футболистов, хоккеистов и штангистов.<br>Даны значения роста в трех группах случайно выбранных спортсменов:<br>Футболисты: 173, 175, 180, 178, 177, 185, 183, 182.<br>Хоккеисты: 177, 179, 180, 188, 177, 172, 171, 184, 180.<br>Штангисты: 172, 173, 169, 177, 166, 180, 178, 177, 172, 166, 170.

```python
fp = np.array([173, 175, 180, 178, 177, 185, 183, 182])
hp = np.array([177, 179, 180, 188, 177, 172, 171, 184, 180])
wl = np.array([172, 173, 169, 177, 166, 180, 178, 177, 172, 166, 170])
```

#### Гипотеза H0 - различия статистически не значимы. Примем уровень значимости а=0.05.

```python
F, p = stats.f_oneway(fp, hp, wl)
F, p
```

* Вывод: Отвергаем нулевую гипотезу на уровне значимости а=0.05

#### Отввет:<br>Различие среднего роста взрослых футболистов, хоккеистов и штангистов является статистически значимым
