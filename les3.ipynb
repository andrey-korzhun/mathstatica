## Урок 3. Описательная статистика. Качественные и количественные характеристики популяции. Графическое представление данных

```python
import numpy as np


def factorial(n):
    if n < 2:
        return 1
    else:
        return n * factorial(n-1)


def combinations(n, k):
    return int(factorial(n) / (factorial(k) * factorial(n - k)))
```

### 1. Даны значения зарплат из выборки выпускников: 100, 80, 75, 77, 89, 33, 45, 25, 65, 17, 30, 24, 57, 55, 70, 75, 65, 84, 90, 150. Посчитать (желательно без использования статистических методов наподобие std, var, mean) среднее арифметическое, среднее квадратичное отклонение, смещенную и несмещенную оценки дисперсий для данной выборки.

```python
salary = [100, 80, 75, 77, 89, 33, 45, 25, 65, 17, 30, 24, 57, 55, 70, 75, 65, 84, 90, 150]
```

**Среднее арифметическое**

```python
total = 0

for x in salary:
    total += x

mean_salary = total / len(salary)
mean_salary
```

**Проверка**

```python
mean_salary == np.mean(salary)
```

**среднее квадратичное отклонение**

```python
total = 0

for x in salary:
    total += (x - mean_salary) ** 2

S = round((total / (len(salary) - 1)) ** 0.5, 2)
S
```

**Проверка**

```python
S == round(np.std(salary, ddof=1), 2)
```

**Смещенная оценка дисперсий**

```python
variance2 = round(total / len(salary), 2)
variance2
```

**Проверка**

```python
variance2 == round(np.var(salary), 2)
```

**Несмещенная оценка дисперсии**

```python
variance2 = round(total / (len(salary) - 1), 2)
variance2
```

**Проверка**

```python
variance2 == round(np.var(salary, ddof=1), 2)
```

**Немного медианы**

```python
def bubble_sort(data):
    lenght = len(data)
    for i in range(lenght):
        swapped = False
        for j in range(0, lenght - i - 1):
            if data[j] > data[j + 1]:
                data[j], data[j + 1] = data[j + 1], data[j]
                swapped = True
        if not swapped:
            break
    print(data)
    return data
```

```python
def my_median(lst):
    n = len(lst)
    return (sum(lst[n//2 - 1: n//2 + 1]) / 2.0, lst[n//2])[n % 2] if n else None
```

```python
median_salary = my_median(bubble_sort(salary))
median_salary
```

**Проверка**

```python
median_salary == np.median(salary)
```

### 2. В первом ящике находится 8 мячей, из которых 5 - белые. Во втором ящике - 12 мячей, из которых 5 белых. Из первого ящика вытаскивают случайным образом два мяча, из второго - 4. Какова вероятность того, что 3 мяча белые?

**Из первого два, из второго один**

```python
C_5_2 = combinations(5, 2)
C_8_2 = combinations(8, 2)
C_5_1 = combinations(5, 1)
C_7_3 = combinations(7, 3)
C_12_4 = combinations(12, 4)
C_5_2, C_8_2, C_5_1, C_7_3, C_12_4
```

**Из первого один, из второго два**

```python
C_3_1 = combinations(3, 1)
C_7_2 = combinations(7, 2)

C_5_1, C_3_1, C_8_2, C_5_2, C_7_2, C_12_4
```

**Из первого ни одного, из второго три**

```python
C_3_2 = combinations(3, 2)
C_5_3 = combinations(5, 3)
C_7_1 = combinations(7, 1)

C_3_2, C_8_2, C_5_3, C_7_1, C_12_4
```

**Итого**

```python
ans_2 = round(C_5_2 / C_8_2 * C_5_1 * C_7_3 / C_12_4 +
              C_5_1 * C_3_1 / C_8_2 * C_5_2 * C_7_2 / C_12_4 +
              C_3_2 / C_8_2 * C_5_3 * C_7_1 / C_12_4, 8)
ans_2
```

### 3. На соревновании по биатлону один из трех спортсменов стреляет и попадает в мишень. Вероятность попадания для первого спортсмена равна 0.9, для второго — 0.8, для третьего — 0.6. Найти вероятность того, что выстрел произведен: a). первым спортсменом б). вторым спортсменом в). третьим спортсменом.

```python
check = 0
P = 0
i = 1
probability = [0.9, 0.8, 0.6]

for shoot in probability:
    P += shoot / len(probability)
    
for biathlete in probability:
    p_i = round(biathlete / len(probability) / P, 4)
    print(f'Спротсмен #{i}: P = {p_i}')
    check += p_i
    i += 1

print(f'Итого: {check:.2f}')
```

### 4. В университет на факультеты A и B поступило равное количество студентов, а на факультет C студентов поступило столько же, сколько на A и B вместе. Вероятность того, что студент факультета A сдаст первую сессию, равна 0.8. Для студента факультета B эта вероятность равна 0.7, а для студента факультета C - 0.9. Студент сдал первую сессию. Какова вероятность, что он учится: a). на факультете A б). на факультете B в). на факультете C?

```python
check = 0
P = 0
i = 0
faculties = ['A', 'B', 'C']
probability = [0.8, 0.7, 2 * 0.9]
for student in probability:
    P += student / len(probability)
for faculty in probability:
    p_i = round(faculty / len(probability) / P, 4)
    print(f'Факультет {faculties[i]}: P = {p_i}')
    check += p_i
    i += 1

print(f'Итого: {check:.2f}')
```

### 5. Устройство состоит из трех деталей. Для первой детали вероятность выйти из строя в первый месяц равна 0.1, для второй - 0.2, для третьей - 0.25. Какова вероятность того, что в первый месяц выйдут из строя: а). все детали б). только две детали в). хотя бы одна деталь г). от одной до двух деталей?

```python
P = [_, 0.1, 0.2, 0.25]
```

**все детали**

```python
round(P[1] * P[2] * P[3], 4)
```

**только две детали**

```python
round(P[1] * P[2] * (1 - P[3]) + P[2] * P[3] * (1 - P[1]) + P[1] * P[3] * (1 - P[2]), 4)
```

**хотя бы одна деталь**

```python
round(1 - (1 - P[1]) * (1 - P[2]) * (1 - P[3]), 4)
```

**от одной до двух деталей**

```python
round(1 - P[1] * P[2] * P[3] - (1 - P[1]) * (1 - P[2]) * (1 - P[3]), 4)
```
