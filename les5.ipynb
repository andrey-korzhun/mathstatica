## Урок 5. Проверка статистических гипотез. P-значения. Доверительные интервалы. A/B-тестирование

```python
import scipy.stats as st
import numpy as np

from IPython.display import display, Markdown
```

```python
def get_t(alpha, n):
    return st.t.ppf(1 - (alpha / 2), n - 1)
```

### 1.Известно, что генеральная совокупность распределена нормально со средним квадратическим отклонением, равным 16. Найти доверительный интервал для оценки математического ожидания с надежностью 0.95, если выборочная средняя M = 80, а объем выборки n = 256.

```python
s = 16
rel = 0.95
M = 80
n = 256
```

```python
z = st.norm.ppf((1 - rel) / 2)
z
```

```python
T_L = M + z * (s / n ** 0.5)
T_R = M - z * (s / n ** 0.5)

display(Markdown(f'**Ответ:**'))
print(f'[{T_L:.4f}, {T_R:.4f}]')
```

### 2. В результате 10 независимых измерений некоторой величины X, выполненных с одинаковой точностью, получены опытные данные: 6.9, 6.1, 6.2, 6.8, 7.5, 6.3, 6.4, 6.9, 6.7, 6.1 Предполагая, что результаты измерений подчинены нормальному закону распределения вероятностей, оценить истинное значение величины X при помощи доверительного интервала, покрывающего это значение с доверительной вероятностью 0,95.

```python
data = [6.9, 6.1, 6.2, 6.8, 7.5, 6.3, 6.4, 6.9, 6.7, 6.1]
rel = 0.95
```

```python
n = len(data)

X = np.mean(data)
s = np.std(data, ddof = 1)

t = get_t(1 - rel, n)
t
```

```python
X = np.mean(data)
s = np.std(data, ddof = 1)

T_L = X - t * (s / n ** 0.5)
T_R = X + t * (s / n ** 0.5)

display(Markdown(f'**Ответ:**'))
print(f'[{T_L:.4f}, {T_R:.4f}]')
```

### 3. Утверждается, что шарики для подшипников, изготовленные автоматическим станком, имеют средний диаметр 17 мм. Используя односторонний критерий с α=0,05, проверить эту гипотезу, если в выборке из n=100 шариков средний диаметр оказался равным 17.5 мм, а дисперсия известна и равна 4 кв.мм. Решать через тестирование гипотезы.

```python
u_0 = 17
u_1 = 17.5
a = 0.05
n = 100
D = 4
s = np.sqrt(D)
```

```python
z_t = st.norm.ppf(1 - a)
z_t
```

```python
z_p = (u_1 - u_0) / (s / n ** 0.5)
z_p
```

```python
display(Markdown(f'**Ответ:**'))
if z_t > z_p:
    print(f'Верна гипотиза H0 на уровне {a:.1%}')
else:
    print(f'Верна гипотиза H1 на уровне {a:.1%}')
```

### 4. Продавец утверждает, что средний вес пачки печенья составляет 200 г. Из партии извлечена выборка из 10 пачек. Вес каждой пачки составляет: 202, 203, 199, 197, 195, 201, 200, 204, 194, 190. Известно, что их веса распределены нормально. Верно ли утверждение продавца, если учитывать, что доверительная вероятность равна 99%? Решать через тестирование гипотезы.

```python
data = [202, 203, 199, 197, 195, 201, 200, 204, 194, 190]
rel = 0.99
a = 1 - rel
u_0 = 200
u_1 = np.mean(data)
u_1
```

```python
n = len(data)

s = np.std(data, ddof = 1)
s
```

```python
T_R = get_t(a, n)
T_R
```

```python
T_L = -get_t(a, n)
T_L
```

```python
T_p = (u_1 - u_0) / (s / n ** 0.5)
T_p
```

```python
display(Markdown(f'**Ответ:**'))
if T_L <= T_p <= T_R:
    print(f'Верна гипотиза H0 на уровне {a:.1%}')
else:
    print(f'Верна гипотиза H1 на уровне {a:.1%}')
```
